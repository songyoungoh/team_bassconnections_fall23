```python
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Fold1",
    "Fold2",
    "Fold3",
    "Fold4",
    "Fold5"
)
weight_counts = {
    "Arceburgo": np.array([a1, a2, a3, a4, a5]),
    "Guaxupe": np.array([guax1, guax2, guax3, guax4, guax5]),
    "Guaranesia": np.array([guar1, guar2, guar3, guar4, guar5]),
    "Monte Santo": np.array([monte1, monte2, monte3, monte4, monte5])
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(5)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of Images from Each State by Fold")
ax.legend(loc="upper right")

plt.show()
```
